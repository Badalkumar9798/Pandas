```python
import numpy as np 
import pandas as pd 
```

**Finding Missing Data**

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
# when you check the null values -> use .isna()
# true -> it means null value is present at that loctation
# false -> it means null value is not present at that location 
```

```python
#working as the columns 
```

```python
df.isna()
```

```python
#when you want to check the total number of null value present in column \\
```

```python
df.isna() .sum() 
```

```python
#check the missing value its means to check all columns are carry null value 
```

```python
df.isna().any()
```
